```python
import numpy as np
import matplotlib.pyplot as plt
import utils
```

Let's start with the simple function we've seen before:

$$ f(x) = 2 \cdot x^2 -3 $$

and plot it

```python
xrange = np.linspace(-1 * np.pi, np.pi, 90)
f = np.fromiter((2 * x * x - 3 for x in xrange), dtype=float)
```

```python
a = plt.subplot(111)
plt.axhline(y=0, color='black')
plt.axvline(x=0, color='black')
a.spines['top'].set_visible(False)
a.spines['right'].set_visible(False)
a.spines['bottom'].set_visible(False)
a.spines['left'].set_visible(False)

a.plot(xrange, f)
a.set_title(r'$2\cdot x^2 = 3$');
a.set_ylim(-5, 15);
```

A natural approach would be to 'bracket' the root. For example, there are two such ranges shown below:

```python
a = plt.subplot(111)
plt.axhline(y=0, color='black')
plt.axvline(x=0, color='black')
a.spines['top'].set_visible(False)
a.spines['right'].set_visible(False)
a.spines['bottom'].set_visible(False)
a.spines['left'].set_visible(False)

a.plot(xrange, f)
a.set_title(r'$2\cdot x^2 + 3 = 0$');
a.set_ylim(-5, 15);

## interval 1: [0,2] and function values are [-3, 5]
x = [0.5, 2]
y = [-2.5, 5]
plt.scatter(x, y, marker='x', color='r');

plt.plot([0.5, 2], [8, 8], color='r', ls='--');
plt.plot([0.5, 0.5], [-2.5, 8], color='grey', ls='dotted');
plt.plot([2, 2], [5, 8], color='grey', ls='dotted');
```

```python
fix, ax = plt.subplots(1, 2, figsize=(16, 5))

for a in ax.flat:
    a.axhline(y=0, color='black')
    a.axvline(x=0, color='black')
    a.spines['top'].set_visible(False)
    a.spines['right'].set_visible(False)
    a.spines['bottom'].set_visible(False)
    a.spines['left'].set_visible(False)
    a.plot(xrange, f)
    a.set_title(r'Bracketing the function $2\cdot x^2 = 3$');
    a.set_ylim(-5, 15);

## interval 1: [0.5,2] and function values are [-2.5, 5]
x = [0.5, 2]
y = [-2.5, 5]
ax[0].scatter(x, y, marker='x', color='r')

ax[0].plot([0.5, 2], [8, 8], color='r', ls='--')
ax[0].plot([0.5, 0.5], [5.8, 8], color='grey', ls='dotted')
ax[0].plot([2, 2], [5.8, 8], color='grey', ls='dotted')
ax[0].text(0.4, 8.6, '0.5', color='orange')
ax[0].text(1.9, 8.6, '2', color='orange')

ax[0].plot([3, 3], [-2.5, 5], color='g', ls='--')
ax[0].plot([2.2, 3], [-2.5, -2.5], color='grey', ls='dotted')
ax[0].plot([2.2, 3], [5, 5], color='grey', ls='dotted')
ax[0].text(3.1, -2.5, '-2.5', color='lime')
ax[0].text(3.1, 5, '5', color='lime')

## interval 2: [-2, 1] and function values are [5, -1]
x = [-2, 1]
y = [5, -1]
ax[1].scatter(x, y, marker='x', color='darkviolet')

ax[1].plot([-2, 1], [8, 8], color='r', ls='--');
ax[1].plot([-2, -2], [6, 8], color='grey', ls='dotted');
ax[1].plot([1, 1], [6, 8], color='grey', ls='dotted');
ax[1].text(-2.1, 8.3, '-2', color='orange');
ax[1].text(0.9, 8.3, '1', color='orange');

ax[1].plot([-2.8, -2.8], [-1, 5], color='g', ls='--');
ax[1].plot([-2.8, -2.2], [-1, -1], color='grey', ls='dotted');
ax[1].plot([-2.8, -2.2], [5, 5], color='grey', ls='dotted');
ax[1].text(-3.1, -1.2, '-1', color='lime');
ax[1].text(-3.1, 4.8, '5', color='lime');
```
