```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

1. range is half-open: (start, stop]
2. r[i] = start + step * i while r[i] < stop if step > 0, and r[i] > stop if step < 0
3. lazy evaluation!!
4. ranges can be compared using '==' or '!=': this compares the sequences they represent
    range(0, 5, 2)
    
see also: numpy.arange and numpy.linspace

linspace Example

```python
num_vectors = 10
radians = np.linspace(0, 2 * np.pi, num_vectors, endpoint=False)
print(radians)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15,15))
print(type(fig))
print(type(ax))

ax.set_title("Exploring the Unit Circle!")
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')

colors = np.linspace(0, 1.0, num_vectors)
print(colors[:5])

cmap = matplotlib.cm.get_cmap('RdBu', num_vectors)
print((type(cmap)))

for i in range(num_vectors):
    xc, yc = np.cos(radians[i]), np.sin(radians[i])
    x0, y0 = [0, xc], [0, yc]
    line = ax.plot(x0, y0, marker='o', color=cmap(colors[i]))
    
plt.show()
```
