Random number generation

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Generate one random sample
s = np.random.random_sample()
print(s)
```

```python
# Sample from a distribution: Normal

s = np.random.normal(5, 2, 1000)
print(type(s))
print(len(s))
print(s[:10])

fig, ax = plt.subplots(1, 1, figsize=(15,10))
ax.set_title("Distribution of samples")

ax.plot(s)

plt.show()
```

```python
# Sample from a distribution: Normal

s = np.random.normal(5, 2, 100000)
print(type(s))
print(len(s))
print(s[:10])

fig, ax = plt.subplots(1, 2, figsize=(10,5))
ax[0].set_title("Distribution of samples")
ax[1].set_title("Histogram of samples")

ax[0].plot(s)
ax[1].hist(s,50)
plt.show()
```
